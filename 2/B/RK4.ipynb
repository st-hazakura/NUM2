```python
import matplotlib.pyplot as plt
import numpy as np
```

## Rungova-Kuttova metoda 4. řádu

```python
def dy(x_n,y_n):
    return (y_n**2+1)/(x_n*y_n)

def RK4(x_0, y_0, x_max, h):
    x, y = [x_0], [y_0]

    while x[-1] < x_max:
        k_1 = h * dy(x[-1], y[-1])
        k_2 = h * dy(x[-1]+ h/2, y[-1] + (h/2) * k_1)
        k_3 = h * dy(x[-1]+ h/2, y[-1] + (h/2) * k_2)
        k_4 = h * dy(x[-1]+ h, y[-1]+ h * k_3)
        y.append(y[-1] + (1/6)*(k_1 + 2*k_2 + 2*k_3 + k_4))    
        x.append(x[-1]+h)
    return x,y
```

## Analyticke reseni

```python
def analyticke_reseni(x_0, y_0, x_max, h):
    y = [np.sqrt(5*i**2 -1) for i in np.arange(x_0, x_max,h)]
    x = [i for i in np.arange(x_0,x_max,h)]
    return x,y
```

## Vysledky

Pocatecni podminky

```python
x_0, y_0 = 1.0, 2.0
x_max = 5.0
h = 0.0001
parametry = [x_0,y_0,x_max,h]
```

```python
r_x,r_y = RK4(*parametry)
a_x,a_y = analyticke_reseni(*parametry)
```

```python
rk_chyba = np.abs(np.array(r_y[:-1]) - np.array(a_y))
```

```python
plt.rcParams.update({'font.size': 12})
plt.title(f"Globální diskretizační chyba h = {h} ")
plt.plot(rk_chyba, label = "RK4 metoda - Analytické řešení")
plt.xlabel("Iterace")
plt.ylabel("Chyba")
plt.legend(loc = 2)
plt.show()
```
